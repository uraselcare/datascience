```python
######## HANdLING OUTLIER ######################
all_data=[np.random.normal(0,std,100)for std in range (5,8)]
fig=plt.figure(figsize=(4,3))
bplot=plt.boxplot(all_data,
            notch=False,
            sym="rs",
            vert=True)
plt.xticks([y+1 for y in range(len(all_data))],["x1","x2","x3"])
plt.xlabel("measurement")
plt.title("Box plot")
print(bplot.keys())
for components in bplot.keys():
  for line in bplot[components]:
    line.set_color("black")
t=plt.title("black and white box plot")
plt.show()
```

```python
all_data=[np.random.normal(0,std,100) for std in range(5,8)]
plt.boxplot(all_data,
            notch=True,
            sym="rs",
            vert=True,
            patch_artist=True)
plt.xticks([y+1 for y in range (len(all_data))],["x1","x2","x3"])
plt.xlabel("measurement")
plt.title("Box plot")
```

```python
all_data=[np.random.normal(0,std,100) for std in range(5,8)]
bplots=plt.boxplot(all_data,
            notch=True,##notch shape
            sym="rs",# outlier color
            vert=True, #vertical
            patch_artist=True)#fill with color
colors=["pink","lightblue","lightgreen"]
for patch,color in zip(bplots["boxes"],colors):
  patch.set_facecolor(color)

plt.xticks([y+1 for y in range (len(all_data))],["x1","x2","x3"])
plt.xlabel("measurement")
plt.title("Box plot")
```

```python
###violin plot
mean_values=[1,2,3]#actual values
variance=[0.2,0.4,0.5]#error associated with each bar
bar_labels=["bar1","bar2","bar3"]
x_pos=list(range(len(bar_labels)))
plt.bar(x_pos,mean_values,yerr=variance,align="center",alpha=0.5)
plt.grid()
max_y=max(zip(mean_values,variance))
plt.ylim([0,((max_y[0]+max_y[1])*1.1)])
```
